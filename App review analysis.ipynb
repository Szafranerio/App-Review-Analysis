```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
linkedin_data = pd.read_csv("/Users/bartlomiejszafran/Desktop/linkedin-reviews.csv")
```

```python
linkedin_data.head()
```

```python
print(linkedin_data.info())
```

```python
sns.pairplot(linkedin_data)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.countplot(data =linkedin_data, x="Rating")
plt.title("Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
#Calculate lenght of the reviews
linkedin_data['Review Lenght'] = linkedin_data ['Review'].apply(len)

#Graph
sns.histplot(linkedin_data['Review Lenght'], bins = 50, kde= True)
plt.title ('Distribution')
plt.xlabel ('Lenght')
plt.ylabel ('Count')
plt.show()
```
